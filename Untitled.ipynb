```python
# WORKING METHOD: Download JSON files from Cricsheet
import requests
from bs4 import BeautifulSoup

def download_cricsheet_files(match_types=['t20', 'odi'], max_files_per_type=10):
    """
    Download JSON files from Cricsheet downloads page
    """
    data_dir = 'cricket_data'
    os.makedirs(data_dir, exist_ok=True)
    
    base_url = "https://cricsheet.org/downloads/"
    downloaded_files = []
    
    print("Fetching file list from Cricsheet...")
    
    for match_type in match_types:
        print(f"\nDownloading {match_type.upper()} matches...")
        
        try:
            # Get the downloads page
            response = requests.get(base_url, timeout=10)
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Find all JSON file links
            json_links = soup.find_all('a', href=True)
            
            count = 0
            for link in json_links:
                href = link['href']
                
                # Check if it's a JSON file for the match type
                if href.endswith('.json') and match_type in href.lower():
                    file_url = base_url + href if not href.startswith('http') else href
                    filename = os.path.join(data_dir, os.path.basename(href))
                    
                    try:
                        # Download the file
                        file_response = requests.get(file_url, timeout=10)
                        if file_response.status_code == 200:
                            with open(filename, 'wb') as f:
                                f.write(file_response.content)
                            print(f"  ✓ Downloaded: {os.path.basename(filename)}")
                            downloaded_files.append(filename)
                            count += 1
                            
                            if count >= max_files_per_type:
                                break
                    except Exception as e:
                        print(f"  ✗ Failed to download {href}: {str(e)}")
                        continue
                        
            print(f"Downloaded {count} {match_type.upper()} files")
            
        except Exception as e:
            print(f"Error fetching {match_type} data: {str(e)}")
    
    return data_dir, downloaded_files

# ALTERNATIVE: Direct ZIP download method
def download_cricsheet_zip(match_type='t20'):
    """
    Download entire match type as ZIP file
    """
    import zipfile
    from io import BytesIO
    
    data_dir = 'cricket_data'
    os.makedirs(data_dir, exist_ok=True)
    
    # Cricsheet provides bulk downloads
    zip_urls = {
        't20': 'https://cricsheet.org/downloads/t20s_json.zip',
        'odi': 'https://cricsheet.org/downloads/odis_json.zip',
        'test': 'https://cricsheet.org/downloads/tests_json.zip'
    }
    
    if match_type not in zip_urls:
        print(f"Unknown match type: {match_type}")
        return data_dir
    
    print(f"Downloading {match_type.upper()} ZIP file...")
    print("This may take a few minutes...")
    
    try:
        response = requests.get(zip_urls[match_type], timeout=60)
        
        if response.status_code == 200:
            # Extract ZIP
            with zipfile.ZipFile(BytesIO(response.content)) as zip_file:
                # Extract only first 20 files for demo
                all_files = zip_file.namelist()
                json_files = [f for f in all_files if f.endswith('.json')][:20]
                
                print(f"Extracting {len(json_files)} files...")
                for file in json_files:
                    zip_file.extract(file, data_dir)
                    print(f"  ✓ Extracted: {file}")
            
            print(f"\n✓ Successfully downloaded {len(json_files)} files to {data_dir}")
            return data_dir
        else:
            print(f"Failed to download. Status code: {response.status_code}")
            return data_dir
            
    except Exception as e:
        print(f"Error downloading ZIP: {str(e)}")
        return data_dir

# TRY METHOD 1: Direct ZIP download (RECOMMENDED - FASTEST)
print("="*60)
print("METHOD 1: Downloading T20 matches (ZIP)")
print("="*60)
data_dir = download_cricsheet_zip('t20')

# If method 1 fails, try downloading ODI as well
if os.path.exists(data_dir):
    json_count = len([f for f in os.listdir(data_dir) if f.endswith('.json')])
    if json_count < 5:
        print("\n" + "="*60)
        print("METHOD 2: Downloading ODI matches (ZIP)")
        print("="*60)
        data_dir = download_cricsheet_zip('odi')

print(f"\nData directory: {os.path.abspath(data_dir)}")
```

```python
# Parse JSON files and create DataFrames
def parse_json_files(directory):
    all_matches = []
    
    # Find all JSON files recursively
    json_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                json_files.append(os.path.join(root, file))
    
    if not json_files:
        print(f"WARNING: No JSON files found in {directory}")
        print("Please run Step 2 to download the data first!")
        return pd.DataFrame()
    
    print(f"Found {len(json_files)} JSON files. Processing...")
    
    for filepath in json_files:
        try:
            with open(filepath, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            if 'info' not in data:
                continue
                
            match_info = data.get('info', {})
            
            match_record = {
                'match_id': os.path.basename(filepath).replace('.json', ''),
                'match_type': match_info.get('match_type', 'Unknown'),
                'gender': match_info.get('gender', 'male'),
                'date': match_info.get('dates', [None])[0] if match_info.get('dates') else None,
                'venue': match_info.get('venue', 'Unknown'),
                'city': match_info.get('city', 'Unknown'),
                'team1': match_info.get('teams', ['Unknown', 'Unknown'])[0] if len(match_info.get('teams', [])) > 0 else 'Unknown',
                'team2': match_info.get('teams', ['Unknown', 'Unknown'])[1] if len(match_info.get('teams', [])) > 1 else 'Unknown',
                'toss_winner': match_info.get('toss', {}).get('winner', 'Unknown'),
                'toss_decision': match_info.get('toss', {}).get('decision', 'Unknown'),
                'winner': match_info.get('outcome', {}).get('winner', 'No Result'),
                'player_of_match': match_info.get('player_of_match', [None])[0] if match_info.get('player_of_match') else None
            }
            
            # Extract innings data
            innings = data.get('innings', [])
            for idx, inning in enumerate(innings[:4]):
                team = inning.get('team', 'Unknown')
                overs = inning.get('overs', [])
                
                total_runs = 0
                total_wickets = 0
                
                for over in overs:
                    for delivery in over.get('deliveries', []):
                        total_runs += delivery.get('runs', {}).get('total', 0)
                        if 'wickets' in delivery:
                            total_wickets += len(delivery['wickets'])
                
                match_record[f'innings{idx+1}_team'] = team
                match_record[f'innings{idx+1}_runs'] = total_runs
                match_record[f'innings{idx+1}_wickets'] = total_wickets
            
            all_matches.append(match_record)
            
        except Exception as e:
            print(f"Error processing {os.path.basename(filepath)}: {str(e)}")
            continue
    
    print(f"\n✓ Successfully processed {len(all_matches)} matches")
    return pd.DataFrame(all_matches)

# Process files
df_all = parse_json_files(data_dir)

if df_all.empty:
    print("\nNo data was processed. Please check Step 2.")
else:
    print(f"\nColumns: {list(df_all.columns)}")
    
    # Split by match type
    df_test = df_all[df_all['match_type'].str.contains('Test|test', case=False, na=False)].copy()
    df_odi = df_all[df_all['match_type'].str.contains('ODI|odi', case=False, na=False)].copy()
    df_t20 = df_all[df_all['match_type'].str.contains('T20|t20', case=False, na=False)].copy()
    df_ipl = df_all[df_all['match_type'].str.contains('IPL|ipl', case=False, na=False)].copy()
    
    print(f"\nMatch Type Distribution:")
    print(f"Test matches: {len(df_test)}")
    print(f"ODI matches: {len(df_odi)}")
    print(f"T20 matches: {len(df_t20)}")
    print(f"IPL matches: {len(df_ipl)}")
    
    print(f"\nUnique match types: {df_all['match_type'].unique()}")
    print(f"\nSample Data:")
    print(df_all.head())
```

```python
# Create SQLite database
engine = create_engine('sqlite:///cricket_matches.db')

df_test.to_sql('test_matches', engine, if_exists='replace', index=False)
df_odi.to_sql('odi_matches', engine, if_exists='replace', index=False)
df_t20.to_sql('t20_matches', engine, if_exists='replace', index=False)
df_ipl.to_sql('ipl_matches', engine, if_exists='replace', index=False)
df_all.to_sql('all_matches', engine, if_exists='replace', index=False)

print("✓ SQLite database created successfully!")

# Verify tables
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, engine)
print("\nTables in database:")
print(tables)

# IMPORTANT: Export to CSV for Power BI
print("\n" + "="*60)
print("EXPORTING FOR POWER BI")
print("="*60)
print("Power BI doesn't support SQLite directly.")
print("Exporting tables to CSV format...")

df_all.to_csv('powerbi_all_matches.csv', index=False)
df_test.to_csv('powerbi_test_matches.csv', index=False)
df_odi.to_csv('powerbi_odi_matches.csv', index=False)
df_t20.to_csv('powerbi_t20_matches.csv', index=False)
df_ipl.to_csv('powerbi_ipl_matches.csv', index=False)

print("\n✓ CSV files created:")
print("  - powerbi_all_matches.csv")
print("  - powerbi_test_matches.csv")
print("  - powerbi_odi_matches.csv")
print("  - powerbi_t20_matches.csv")
print("  - powerbi_ipl_matches.csv")
print("\nUse these CSV files in Power BI: Get Data > Text/CSV")
print("="*60)
```

```python
# 20 SQL Queries for Analysis
queries = {
    "Q1": "SELECT match_type, COUNT(*) as total FROM all_matches GROUP BY match_type",
    "Q2": "SELECT venue, COUNT(*) as matches FROM all_matches GROUP BY venue ORDER BY matches DESC LIMIT 10",
    "Q3": "SELECT winner, COUNT(*) as wins FROM all_matches WHERE winner != 'No Result' GROUP BY winner ORDER BY wins DESC LIMIT 10",
    "Q4": "SELECT SUM(CASE WHEN toss_winner = winner THEN 1 ELSE 0 END) as toss_advantage FROM all_matches WHERE winner != 'No Result'",
    "Q5": "SELECT toss_decision, COUNT(*) as count FROM all_matches GROUP BY toss_decision",
    "Q6": "SELECT match_type, AVG(innings1_runs) as avg_runs FROM all_matches GROUP BY match_type",
    "Q7": "SELECT match_id, venue, (innings1_runs + innings2_runs) as total FROM all_matches ORDER BY total DESC LIMIT 10",
    "Q8": "SELECT strftime('%Y', date) as year, COUNT(*) as matches FROM all_matches WHERE date IS NOT NULL GROUP BY year",
    "Q9": "SELECT player_of_match, COUNT(*) as awards FROM all_matches WHERE player_of_match IS NOT NULL GROUP BY player_of_match ORDER BY awards DESC LIMIT 10",
    "Q10": "SELECT city, COUNT(*) as matches FROM all_matches WHERE city != 'Unknown' GROUP BY city ORDER BY matches DESC LIMIT 10"
}

results = {}
for name, query in queries.items():
    print(f"Executing: {name}")
    results[name] = pd.read_sql(query, engine)
    print(results[name])
    print("-" * 60)
```
